### Problem 6: Oscillations and Bistability in the Morris-Lecar Model

### Part A

We want to simulate the Morris-Lecar model given by:
$$ C \frac{du}{dt} = -g_1m_0(u) (u - V_1)-g_2\hat{w}(u-V_2)-g_L(u-V_L) + I $$
$$ \frac{d\hat{w}}{dt} = -\frac{1}{\tau(u)}[\hat{w} - w_0(u)] $$

where $$ m_0(u) = \frac{1}{2}[1+\text{tanh}(\frac{u-u_1}{u_2})] $$ $$ w_0(u) = \frac{1}{2}[1+\text{tanh}(\frac{u-u_3}{u_4})] $$ We can also approximate the time constant, $\tau$ as: $$ \tau(u)=\frac{\tau_w}{\text{cosh}(\frac{u-u_3}{2u_4})} $$ 

```python
# set up constants
tau_w = 25
g_1 = 4.4
g_2 = 8
g_L = 2
V_1 = 120
V_2 = -84
V_L = -60
u_1 = -1.2
u_2 = 18
u_3 = 2
u_4 = 30
C = 20
I = 120
```

```python
# define some helper functions
def m_0(u):
    return 0.5*(1+np.tanh((u-u_1)/u_2))
def w_0(u):
    return 0.5*(1+np.tanh((u-u_3)/u_4))
def tau(u):
    return tau_w/np.cosh((u-u_3)/(2*u_4))
```

```python
def differentials(y, time):
    dy = np.zeros((2,))
    
    u = y[0]
    w_hat = y[1]
    
    dy[0] = -g_1*m_0(u)*(u-V_1)-g_2*w_hat*(u-V_2)-g_L*(u-V_L) + I
    dy[0] /= C
    
    dy[1] = (-1/tau(u))*(w_hat-w_0(u))
    
    return dy
```

```python
def plot_result(tvec, result):
    fig = plt.figure()
    plt.grid(True, alpha=0.2)
    plt.plot(tvec, result[:, 0])
    plt.xlabel('time (ms)')
    plt.ylabel('voltage (mV)')
    plt.title('u')
    plt.show()

    fig = plt.figure()
    plt.grid(True, alpha=0.2)
    plt.plot(tvec, result[:, 1])
    plt.xlabel('time (ms)')
    plt.ylabel('recovery amount')
    plt.title('w')
    plt.show()

    fig = plt.figure()
    plt.grid(True, alpha=0.2)
    plt.plot(result[:,1], result[:, 0])
    plt.xlabel('w')
    plt.ylabel('u')
    plt.title('Phase Plane')
    plt.show()
```

```python
from scipy.integrate import odeint
```

```python
T = 500.0    # end time (in milliseconds)
tvec = np.linspace(0, T, 10000)    # time vector
initial = np.array([0,0])
result = odeint(differentials, initial, tvec)

# plot result
plot_result(tvec, result)
```

The time between peaks is approximately 73 ms.
